```python
import pandas as pd

df1 = pd.read_csv('Dataset/WHO-COVID-19-global-data.csv', sep=';')
df2 = pd.read_csv('Dataset/population_by_country_2020.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1.drop(['WHO_region','New_cases','New_deaths','Country_code'], axis=1, inplace=True)
df1.head()
```

```python
df2 = df2[['Country (or dependency)', 'Population (2020)']]
df2.head()
```

```python
df2.rename(columns={'Country (or dependency)': 'Country', 'Population (2020)': 'Population'}, inplace=True)
df2.head()
```

```python
df = pd.merge(df1, df2, left_on='Country', right_on='Country', how='inner')

df['Cumulative_cases_per_100k'] = (df['Cumulative_cases'] / df['Population']) * 100000
df['Cumulative_deaths_per_100k'] = (df['Cumulative_deaths'] / df['Population']) * 100000

df.head()
```

```python
df = df[['Date_reported', 'Country', 'Cumulative_cases_per_100k', 'Cumulative_deaths_per_100k']]
df.head()
```

```python
df.to_json('df.json', orient='records', lines=True)
```

```python
import re

# Leer el archivo JSON
with open('df.json', 'r') as file:
    content = file.read()

# Agregar las comas que faltan entre los objetos JSON
# Busca patrones de objetos JSON y agrega una coma entre ellos
corrected_content = re.sub(r'(?<=\})(\s*)(?=\{)', r',\1', content)

# Envolver el contenido en corchetes para formar un array JSON válido
corrected_content = f'[{corrected_content}]'

# Guardar el contenido corregido en un nuevo archivo
with open('df.json', 'w') as file:
    file.write(corrected_content)

print("Archivo corregido guardado como 'df.json'")
```
